Webscraping and Pandas

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue'

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36"
}

page = requests.get(url, headers=headers)

# Retrieves HTML from specified webpage
soup = BeautifulSoup(page.text, 'html')
```

```python
# Retrieves the specific data table desired on webpage
table = soup.find('table', class_ = 'wikitable')
```

```python
# Retrieves headers (columns) from table
world_titles = table.find_all("th")

# Converts headers to text
world_table_titles = []
for title in world_titles:
    world_table_titles.append(title.get_text(strip=True))

# Create dataframe in panndas
df = pd.DataFrame(columns=world_table_titles)

# Retrieves rows from table
table_rows = table.find_all("tr")

# Retrieves data entries from rows (skips header row)
for row in table_rows[1:]:
    row_data = row.find_all("td")
    
    individual_row_data = []
    for data in row_data:
        individual_row_data.append(data.get_text(strip=True))

    # Loops through each row of dataframe and adds row data
    length = len(df)
    df.loc[length] = individual_row_data

df
```
